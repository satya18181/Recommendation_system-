```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"},{"id": 12, "name": "Adventure"},{"id": 80, "name": "Crime"}]')
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies1['genres']=movies1['genres'].apply(convert)
movies1['keywords']=movies1['keywords'].apply(convert)
movies1['cast']=movies1['cast'].apply(convert)
movies1['crew']=movies1['crew'].apply(convert)
```

```python
movies1
```

```python
movies1.isnull().sum()/len(movies1)*100
```

```python
movies1['title'].nunique()
```

```python
movies1.dropna(inplace=True)
```

```python
movies1
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies1['genres'] = movies1['genres'].apply(collapse)
movies1['keywords'] = movies1['keywords'].apply(collapse)
movies1['cast'] = movies1['cast'].apply(collapse)
movies1['crew'] = movies1['crew'].apply(collapse)
```

```python
movies1
```

```python
movies1['overview'] = movies1['overview'].apply(lambda x: x.split())
```

```python
movies1.head(2)
```

```python
movies1['context'] = movies1['genres'] + movies1['keywords'] + movies1['overview'] + movies1['cast'] + movies1['crew']
```

```python
movies1.head(2)
```

```python
movies1
```

```python
new_movie = movies1.drop(columns=['genres','keywords','overview','cast','crew'], axis=1)
```

```python
new_movie
```

```python
new_movie['context'] = new_movie['context'].apply(lambda x: " ".join(x))
```

```python
new_movie
```

```python
new_movie['context'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector = cv.fit_transform(new_movie['context']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
new_movie[new_movie['title']=='Avatar'].index[0]
```

```python
def recommendation(movies):
    index = new_movie[new_movie['title']==movies].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True,
                       key = lambda x :x[1])
    for i in distances[1:6]:
        print(new_movie.iloc[i[0]].title)
```

```python
recommendation('Avatar')
```

```python
recommendation('Spider-Man 3')
```

```python
recommendation('Captain America: The First Avenger')
```
